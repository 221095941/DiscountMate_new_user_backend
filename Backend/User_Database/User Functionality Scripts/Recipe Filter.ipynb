# Recipe and Allergy Filter Script

This script takes a csv with a list of ingredients and allows the user to filter based on desired ingredients.

Possilbe expansions:
1. Add tags to recipes to allow for easy filtering for dietary preference.
2. Easier to chain requests together to narrow down searches and boolean functionality.
3. Integrate better with the web scraping team as we need category data as well.

```python
# Test CSV (Title/Ingredients) - 100 rows in base64 for portability
# Full 1.2GB dataset available at - https://recipenlg.cs.put.poznan.pl/dataset

recipes_csv_64 = ""
```

```python
import base64
import csv
from io import StringIO
import pandas as pd

decoded_data = base64.b64decode(recipes_csv_64)

csv_string = decoded_data.decode('utf-8')

data = pd.read_csv(StringIO(csv_string))

print(data.head(10))
```

```python
# Define the search ingredient
search_ingredient = 'sugar'

# Filter the DataFrame
filtered_data = data[data['ingredients'].apply(lambda x: search_ingredient in x)]['title']

# Print the filtered DataFrame
print(filtered_data)
```
