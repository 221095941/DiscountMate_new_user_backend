# Fuzzy Search on Scraped Data

The following is a utility script which takes a csv column as input, the item name, and a query and returns the top 5 matches from the given column. 

```python
import base64
import csv
from io import StringIO
import pandas as pd
from fuzzywuzzy import process

decoded_csv = base64.b64decode(woolworths_csv_64)

csv_string = decoded_csv.decode('utf-8')

data = pd.read_csv(StringIO(csv_string))

print(data.head(10))
```

```python
# String to search for
search_query = 'Apple'

# Extract first column if csv returned in full
first_column = data.columns[0]

# Find the 5 closest matches
matches = process.extract(search_query, data[first_column], limit=5)

# Print the matches 
# Format of returned tuple - (Item name, Match Score, Rank)
print(matches)

top_match = matches[0]
print("\nThe top match is '" + top_match[0] + "' with a match score of " + str(top_match[1]))
```
